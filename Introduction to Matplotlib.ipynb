### Matplot example workflow

```python
# 0. import matplot and get it ready for plotting
# inline: plot in jupyter notebook
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup Plot
fig, ax = plt.subplots(figsize=(10, 10)) # (width, height)

# 3. Plot data
ax.plot(x, y)

# 4. Customize Plot
ax.set(title="Simple Plot",
      xlabel="x-axis",
      ylabel="y-axis")

# 5. Save & show
fig.savefig("./images/sample-plot.png")
```

## Making figures with numpy arrays
 
 we want:
 * Line plot
 * Scatter plot
 * Bar plot
 * Histogram plot
 * Subplots

```python
import numpy as np
```

```python
# create some data
x = np.linspace(0, 10, 100)
x[:10]
```

```python
# Plot the data and create a line plot
fig, ax = plt.subplots()
ax.plot(x, x**2)
```

```python
# Plot the data and create a scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x))
```

```python
# Another Scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x))
```

```python
# Make a plot from a dictionary
nut_butter_prices = {"Almond Butter": 10,
                    "Peanut Butter": 8,
                    "Cashew Butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title="Dan's Nut butter store",
      ylabel="Prices($)"); # ; to remove unwanted output
```
